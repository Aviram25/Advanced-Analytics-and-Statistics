```python
import numpy as np

arr = np.arange(10)
print(arr)
```

```python
print(arr + arr)
print(arr * arr)
print(arr - arr)
print(arr / arr)
print(arr ** 2)
print(np.sqrt(arr))
```

```python
print(arr.sum())
print(arr.mean())
print(arr.std())
print(arr.max())
print(arr.min())
print(arr.var())
```

```python
students = np.array([
    ('Alice', 85, 20),
    ('Bob', 90, 22),
    ('Charlie', 78, 19)
],   dtype=[('name', 'U10'), ('marks', 'i4'), ('age', 'i4')])

print(students['name'])
print(students['marks'])
print(students['marks'].mean())
print(students[students['name'] == 'Bob']['marks'])
print(students[students['name'] == 'Bob']['marks'][0])
```

```python
sales = [0,5,155,0,518,616]
sales_array = np.array(sales)

print(type(sales_array))
print(sales_array.dtype)
print(sales_array.shape)
print(sales_array.ndim)
print(sales_array.size)
```

```python
result = np.where(sales_array == 0, 'No Sales', 'Sales Made')
print(result)
```

```python
bool_mask = sales_array != 0
print(bool_mask)
```

```python
filtered_sales = sales_array[bool_mask]
print(filtered_sales)
```

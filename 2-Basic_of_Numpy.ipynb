```python
marks = np.array([55,68,49,83,91,37,65,78,59,88])
top_batch = marks[0:5]
print(top_batch)
```

```python
top_batch[:] = 99
print(top_batch)
print(marks)
```

```python
marks = np.array([55,68,49,83,91,37,65,78,59,88])
marks_copy = marks.copy()
marks_copy[3] = 100
print(marks)
print(marks_copy)
```

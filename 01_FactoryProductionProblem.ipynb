```python
# LLP solution using PuLP
from pulp import  *
```

```python
# Create the problem variable to contain the problem data
model = LpProblem("FactoryProductionProblem",LpMaximize)
```

```python
# Create variables for Chairs and Tables
# Parameter: Name, Lower limit, Upper limit, Data type
chairs = LpVariable("Chairs",0,None,LpInteger) # Number of chairs to produce
tables = LpVariable("Tables",0,None,LpInteger) # Number of tables to produce
```

```python
# Create maximize objective function (profit)
model += 300*chairs + 1000*tables,"Profit"
```

```python
# Create constraints
model += 5*chairs + 10*tables <= 400 , "Wood Constraint", # Wood availability constraints
model += 2*chairs + 6*tables <= 300, "Time Constraint" # Wood availability constraint
model += chairs >=2," Chairs Constraint"
```

```python
# The problem is solved using PuLP's choice of Solver
model.solve()
```

```python
# Each of the variables is printed with its resolved optimum values 
for v in model.variables():
    print(v.name,"=",v.varValue)
```

```python
# Additional Information 
print("Total Profit:",value(model.objective)) # Print function value (Total Profit)
```

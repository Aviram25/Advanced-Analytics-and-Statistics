```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
commuter_times = [16,8,35,17,13,15,15,5,16,25,20,20,12,10]
data_arr = np.array(commuter_times)

```

```python
q1  = np.percentile(data_arr,25)
q3  = np.percentile(data_arr,75)
iqr = q3-q1
lower_fence = q1-(1.5*iqr)
upper_fence = q3+(1.5*iqr)

outliers = data_arr[(data_arr<lower_fence) | (data_arr>upper_fence)]
print(outliers)
```

```python
print(f'Q1 : {q1}')
print(f'Q3 : {q3}')
print(f'IQR: {iqr}')
print(f'LOWER_FENCE:  {lower_fence}')
print(f'UPPER_FENCE:  {upper_fence}')
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=data_arr, orient='h')
plt.title('Boxplot of Commuter Times')
plt.xlabel('Minutes')
plt.show()
```
